```python
import requests #파이썬 Requests 은 웹 서버에 HTTP 요청을 하기 위한 라이브러리
import pprint
import json
```

```python
#api 불러오기, url주소 붙여넣고 개인 인증키(decoding)복붙 후 페이지 행의 수 타입을 &로 연결하여 보여준다.
 # url = "http://api.data.go.kr/openapi/tn_pubr_public_ovrspd_prvn_manage_api?serviceKey=vUC2Hcod8KJDx5P6TcpXMXvTDsjDvtpMWpbfnr03F/yuvFWRjEWgFMOUDBXVtCHUQ4lHzYkI1S20+BpRVrW7Zw==&pageNo=1&numOfRows=1000&type=json"

url = "http://api.data.go.kr/openapi/tn_pubr_public_ovrspd_prvn_manage_api?serviceKey=vUC2Hcod8KJDx5P6TcpXMXvTDsjDvtpMWpbfnr03F%2FyuvFWRjEWgFMOUDBXVtCHUQ4lHzYkI1S20%2BBpRVrW7Zw%3D%3D&pageNo=1&numOfRows=1000&type=json"
```

```python
response = requests.get(url)
contents = response.text

## 예ㅃ그게 출력하기 위해
## indent는 들여쓴다는 의미

pp =pprint.PrettyPrinter(indent=4)

print(pp.pprint(contents))

```

```python
json_ob = json.loads(contents)
print(pp.pprint(json_ob))
```

```python
body = json_ob['response']['body']['items']
print(body)
```

```python
import pandas as pd
from pandas.io.json import json_normalize

dataframe = json_normalize(body)
print(dataframe)

dataframe.to_csv('과속방지턱.csv',index = False)
```
